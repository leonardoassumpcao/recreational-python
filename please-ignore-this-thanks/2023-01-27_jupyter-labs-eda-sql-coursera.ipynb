```python
# Save csv as sqlite database:
import pandas as pd
import sqlite3

# Read csv file, ignoring header:
df = pd.read_csv('./Spacex.csv', header=0)
df.columns = "event_date event_time booster_version launch_site payload payload_mass_kg orbit customer mission_outcome landing_outcome".split()

df.head(2)
```

```python
# Create sqlite database:
conn = sqlite3.connect('spacex.db')
c = conn.cursor()

c.execute("DROP TABLE IF EXISTS SPACEX;")

# Create table:
c.execute('''CREATE TABLE SPACEX (
    event_date DATE NOT NULL,
    event_time TIME,
    booster_version VARCHAR(20),
    launch_site VARCHAR(20),
    payload VARCHAR(80),
    payload_mass_kg SMALLINT,
    orbit VARCHAR(20),
    customer VARCHAR(80),
    mission_outcome VARCHAR(40),
    landing_outcome VARCHAR(40)
);''')
```

```python
# # Show a few rows
# for i, row in enumerate(df.itertuples()):
#     print(*map(repr, row), end="\n\n")
#     if i > 3:
#         break

# Insert data into table:
for row in df.itertuples():
    c.execute("INSERT INTO SPACEX VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?);", row[1:])

# Commit changes:
conn.commit()
```

## Tasks

Now write and execute SQL queries to solve the assignment tasks.


### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
%%sql
-- Task 1: Display the names of the unique launch sites in the space mission:

-- "Explanation: The DISTINCT keyword is used to return only distinct (different) values."

SELECT DISTINCT launch_site
FROM SPACEX;
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
%%sql
-- Task 2: Display 5 records where launch sites begin with the string 'CCA':

-- "Explanation: The LIKE operator is used to search for a specified pattern in a column. LIMIT is used to limit the number of records returned."

SELECT * FROM SPACEX
WHERE launch_site LIKE 'CCA%'
LIMIT 5;
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
%%sql
-- Task 3: Display the total payload mass carried by boosters launched by NASA (CRS):

-- "Explanation: COUNT(DISTINCT [...]) = number of rows matching criteria; SUM() = total sum of a numeric column."

SELECT
    COUNT(DISTINCT customer) AS 'nasa_unique_customer_count',
    SUM(payload_mass_kg) AS 'total_payload_mass_kg'
FROM SPACEX
WHERE customer LIKE '%NASA%';
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
%%sql
-- Task 4: Display average payload mass carried by booster version F9 v1.1:

-- "Explanation: AVG() = average of a numeric column; ROUND() = round a number to a specified number of decimals."

SELECT
    booster_version,
    ROUND(AVG(payload_mass_kg), 3) AS avg_payload_mass_kg
FROM SPACEX
WHERE booster_version LIKE 'F9 v1.1%';
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
%%sql
-- Task 5: List the date when the first successful landing outcome in ground pad was acheived:

-- "Explanation: ORDER BY is used to sort the result-set (by default, in ascending order). LIMIT limits the number of records returned."

SELECT
    event_date,
    landing_outcome
FROM SPACEX
WHERE landing_outcome LIKE 'Success%'
ORDER BY event_date
LIMIT 1;
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
%%sql
-- Task 6: List the names of the boosters which have success in drone ship
-- and have payload mass greater than 4000 but less than 6000:

-- "Explanation: BETWEEN is used to select values within a given range. The values can be numbers, text, or dates."

SELECT
    booster_version,
    landing_outcome,
    payload_mass_kg
FROM SPACEX
WHERE landing_outcome = 'Success (drone ship)'
AND payload_mass_kg BETWEEN 4000 AND 6000;
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
%%sql
-- Task 7: List the total number of successful and failure mission outcomes:

-- We must match "Success" as well as "Success (payload status unclear)".
-- In DB2, we could use a simple IF(cond, result_if_true, result_if_false).
-- In Sqlite (our `case` here), we do as follows:

-- "Explanation: CASE is used to create different outputs (usually in the SELECT statement). It is SQLite's way of handling if-then logic."
-- "Alternative: In DB2, we would use IF(mission_outcome LIKE 'Success%', 'SUCCESS', 'FAILURE'), but it is not supported by Sqlite."

SELECT
    CASE
        WHEN mission_outcome LIKE 'Success%'
        THEN 'SUCCESS'
        ELSE 'FAILURE'
    END AS outcome,
    COUNT(*) AS 'count'
FROM SPACEX
GROUP BY outcome;
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
%%sql
-- Task 8: List the names of the booster_versions which have carried the maximum payload mass. Use a subquery.

-- "Explanation: A subquery is a query within a query. The subquery is executed first, and the main query uses its result."

SELECT
    booster_version,
    payload_mass_kg
FROM SPACEX
WHERE payload_mass_kg = (
    SELECT MAX(payload_mass_kg) FROM SPACEX
);
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
%%sql
-- Task 9: List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015:

-- "Explanation: Again, the BETWEEN operator is used to select values within a range. The values can be numbers, text, or dates."

SELECT
    event_date,
    landing_outcome,
    booster_version,
    launch_site
FROM SPACEX
WHERE landing_outcome = 'Failure (drone ship)'
AND event_date BETWEEN '2015-01-01' AND '2015-12-31';
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
%%sql
-- Task 10: Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad))
-- between the date 2010-06-04 and 2017-03-20, in descending order

-- "Explanation: GROUP BY = group rows that have the same values; COUNT(*) = number of rows per group; ORDER BY = sort the result-set."

SELECT
    landing_outcome,
    COUNT(*) AS 'count' FROM SPACEX
WHERE event_date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY landing_outcome
ORDER BY count DESC;
```
